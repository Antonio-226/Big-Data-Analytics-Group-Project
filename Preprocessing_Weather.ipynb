```python
from collections import Counter
problematic_cols = []
for j in df_weather.columns:
    
    ele_list = [] #instantiate list for elements in each column that are not of class int or float

    type_list = [type(i) for i in df_weather[j]]
    type_dict = Counter(type_list)
    if len(type_dict) > 1:
        ele_list = [i for i in df_weather[j] if type(i) == str] #list that contains strings in order to check for "false" classes
        problematic_cols.append(j)
    print(type_dict)
    print('elements with class string of column {}'.format(j))
    print(Counter(ele_list))
    print('')
```

From the result of the code above, wee see that there seems to be a problem with some values that are actually defined but have a different data type than the other non-empty rows and they are actually of the same type. 

```python
#define helper functions to extract floats and integers from strings:
def RepresentsFloat(s):
    try: 
        float(s)
        return True
    except ValueError:
        return False
    
def RepresentsInt(s):
    try:
        int(s)
        return True
    except ValueError:
        return False

    

for j in problematic_cols:
    
    #This loop for each column checks whether its entries are of the same type as its first entry
    #(which is assumed to be the actual/correct data type) and if ith row entry is a string it checks if it is supposed
    #to be a float and if positive then converts the string to a float. The same is done if the ith row is supposed to be 
    #an integer.
    
    for i in range(len(df_weather)):
        
        #if the ith row of column j is a string that contains a float and is supposed to be a float -> convert to float
        if ((type(df_weather[j].values[i]) == str) & (RepresentsFloat(df_weather[j].values[i])) & (type(df_weather[j].values[0]) == float)):
            df_weather[j].values[i] = float(df_weather[j].values[i])
        
        #if the ith row of column j is a string that contains an integer and is supposed to be an integer -> convert to integer
        elif ((type(df_weather[j].values[i]) == str) & (RepresentsInt(df_weather[j].values[i]))):
            df_weather[j].values[i] = int(df_weather[j].values[i])
    
    
```

```python
# Create the empty dataframe for the hourly consolidated data

columns = ['date', 'air_temperature', 'water_temperature', 'wind_gust_max_10min',
           'wind_speed_avg_10min', 'wind_force_avg_10min', 'wind_direction',
           'windchill', 'barometric_pressure_qfe', 'precipitation', 'dew_point',
           'global_radiation', 'humidity', 'water_level']

df_hourly_weather = pd.DataFrame(columns = columns)
```

```python
# Initialize first date in the dataframe

stamp = df_weather['date'][0]

# Initialize counter that keeps track of how many measurements are made in an hour to calculate averages

counter = 0

# Set all other variables to 0

air_temp = 0
water_temp = 0
wind_gust_max = 0
wind_speed_avg = 0
wind_force_avg = 0
wind_direction = 0
windchill = 0
pressure = 0
precipitation = 0
dew_point = 0
radiation = 0
humidity = 0
water_level = 0

for index, row in df_weather.iterrows():
    
    # Compare the date and the hour of the saved variable stamp with this row
    
    if stamp.date() == row['date'].date() and stamp.hour == row['date'].hour:
        counter += 1
        air_temp = air_temp + row['air_temperature']
        water_temp = water_temp + row['water_temperature']
        wind_gust_max = max(wind_gust_max, row['wind_gust_max_10min'])
        wind_speed_avg = wind_speed_avg + row['wind_speed_avg_10min']
        wind_force_avg = wind_force_avg + row['wind_force_avg_10min']
        wind_direction = wind_direction + row['wind_direction']
        windchill = windchill + row['windchill']
        pressure = pressure + row['barometric_pressure_qfe']
        precipitation = precipitation + row['precipitation']
        dew_point = dew_point + row['dew_point']
        radiation = radiation + row['global_radiation']
        humidity = humidity + row['humidity']
        water_level = max(water_level, row['water_level'])
    else:
        df_hourly_weather = df_hourly_weather.append({'date' : datetime(stamp.year, stamp.month, stamp.day, stamp.hour, 30),
                        'air_temperature' : air_temp / counter,
                        'water_temperature' : water_temp / counter,
                        'wind_gust_max_10min' : wind_gust_max,
                        'wind_speed_avg_10min' : wind_speed_avg / counter,
                        'wind_force_avg_10min' : wind_force_avg / counter,
                        'wind_direction' : wind_direction,
                        'windchill' : windchill / counter,
                        'barometric_pressure_qfe' : pressure / counter,
                        'precipitation' : precipitation / counter,
                        'dew_point' : dew_point / counter,
                        'global_radiation' : radiation / counter,
                        'humidity' : humidity / counter,
                        'water_level' : water_level / counter}, ignore_index = True)
        

        # Reset counter and increase the stamp variable with the dateTime of the next row in the df
        
        counter = 0
        stamp = df_weather.iloc[index+1]['date']
        
        # Reset varibles
        
        air_temp = 0
        water_temp = 0
        wind_gust_max = 0
        wind_speed_avg = 0
        wind_force_avg = 0
        wind_direction = 0
        windchill = 0
        pressure = 0
        precipitation = 0
        dew_point = 0
        radiation = 0
        humidity = 0
        water_level = 0
```
